```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

S'ha transformat el fitxer txt per un csv. 

Es carreguen les dades del fitxer.

```python
def load_data(path):
    dataset = pd.read_csv(path, header=0, delimiter=';')
    return(dataset)
```

```python
demographic = load_data('20211025_ADNI_demographics.csv')
```

Un cop eliminades, s'ha de comrpovar que totes les columnes del dataset són de la classe adient, és a dir, que les columnes que conten valor de tipus int, siguin de classe int, etcètera.

```python
for i in demographic.columns:
    print(i, ':', demographic[i].dtypes)
```

Es veu fàcilment com la columna _RID_, que és un identificador, i la columna _APOE4_, que indica el nombre de gens actuis referents a l'Alzheimer de cada cromosoma, haurien de ser de tipis int, però, el primer és un objecte i el segon s'ha carregat com a float. Per tant, es canvia el tipus de classe d'aquestes dues columnes.

Per estalviar errors a l'hora de canviar el tipus de dades, s'ignoraran i es canviaran més tard. A més, els NA de la columna _APOE4_ s'omplen amb un valor de 99.

```python
# Es canvia el type de les columnes 'RID' i 'APOE4'
for i in range(demographic.shape[0]):
    demographic.loc[i,'RID'] = pd.to_numeric(demographic.loc[i,'RID'], errors='ignore')

demographic['APOE4'] = demographic['APOE4'].fillna(99).astype(np.int64)
```

```python
# Es comprova que tots els valors de les columnes s'hagin canviat correctament 
rid_errors = []
apoe_errors = []

for i in range(demographic.shape[0]):
    if type(demographic.loc[i, 'RID']) != np.int64:
        rid_errors.append(i)
    if type(demographic.loc[i, 'APOE4']) != np.int64:
        apoe_errors.append(i)

print('rid errors: ', len(rid_errors))
print('apoe 4 errors: ', len(apoe_errors))
```

```python
# Es mira com són les dades que causen errors
demographic['RID'][rid_errors]
```

Totes les files que no han canviat de tipus, tenen dates com a valors, s'observa com són els altres valors de la fila, en cas que les dades de les altres columnes siguin irrellevants, s'eliminen les files.

```python
demographic.iloc[rid_errors]
```

```python
# Les files no aporten informació. Es podem eliminar
demographic = demographic.drop(rid_errors, axis=0)

#Es resetegen els índex de les fles
demographic = demographic.reset_index(drop = True)

# S'acava de definir la classe de la columna sense errors
demographic['RID'] = pd.to_numeric(demographic['RID'])
```

Un cop netejat el dataset, ja es pot treballar amb ell. Per omplir els _missings_ que falten, primer s'ha de mirar si el pacient apareix més cops dins de la base de dades. Per fer això, s'ha de crear un ID propi de cada pacient. Aquest ID, surt dels 4 últims dígits dels valors de la columns _PTID_. Abans d'extreure aquest ID, s'ha de comprovar que tots els _PTID_ segueixen el mateix patró.

```python
id_error = []
for i in range(demographic.shape[0]):
    check = bool(re.match(r'[0-9][0-9][0-9]_S_[0-9][0-9][0-9][0-9]+', str(demographic['PTID'][i])))
    if not check:
        id_error.append(i)
        
print('id_error: ', len(id_error))
```

Un cop vist que tots els _PTID_ segueixen el mateix patró, ja es pot extreure el ID.

```python
# Es transforma la columna 'PTID' en un identificador de tipus int quedant-nos només amb els 4 últims dígits del str
for i in range(demographic.shape[0]):
    demographic.loc[i,'PTID'] = demographic.loc[i,'PTID'].split('_')[2]

demographic
```

Es mira si es té la informació necessària per omplir els _missings_ que falten. Però primer es comprova si cap pacient té _missings_ a les dues columnes torbades, _DX.bl_ i _AGE_.

```python
# Es busca i guarda el índex dels pacients amb missings a DX.bl
dx = demographic[demographic['DX.bl'].isna()].index.to_list()

# Es busca i guarda el índex dels pacients amb missings a AGE
age = demographic[demographic['AGE'].isna()].index.to_list()

print(age[0] in dx)
print(age[1] in dx)
```

```python
# Es busca i guarda el PTID dels pacients amb missings a DX.bl
idxs = demographic[demographic['DX.bl'].isna()]['PTID'].to_list()

# Es comprova si el pacient apareix més d'un cop a la base de dades
pacient = [i for i in idxs if len(demographic[demographic['PTID'] == i]) == 1]

# Nombre de pacients amb missing
print('pacients incomplets: ', len(pacient))
```

```python
# Es repeteix el procediment anterioir per als missings a la columna AGE

# Es busca i guarda el PTID dels pacients amb missings a AGE
idxs = demographic[demographic['AGE'].isna()]['PTID'].to_list()

# Es comprova si el pacient apareix més d'un cop a la base de dades
pacient = [i for i in idxs if len(demographic[demographic['PTID'] == i]) == 1]

# Nombre de pacients amb missing
print('pacients incomplets: ', len(pacient))
```
